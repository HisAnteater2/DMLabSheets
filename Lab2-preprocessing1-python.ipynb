Q-1:Read and show the properties of dataset "iris_data_with_missing_values.csv"

```python
#Added column names to iris_data_with_missing_values.csv (Please find the file attached with this submission)
#Solution to Q-1
import pandas as pd
import numpy as np
df = pd.read_csv('iris_data_with_missing_values.csv')

print(df.to_string())
```

Q-3: Find the number of zeros in all the columns. 

```python
 #Solution to Q-3
print((df == 0).sum().sum())
```

Q-4: Find the number of null values in the dataset. 

```python
#Solution to Q-4
print(df.isnull().sum().sum())
```

Q-5: Replace the zeros with NAN values in selected attribute (say attribute 1 and 3)

```python
#Solution to Q-5
df[['sepal length in cm', 'petal length in cm']]=df[['sepal length in cm', 'petal length in cm']].replace(to_replace=0, value=np.NaN)
print (df)
```

 Q-6: Find the number of rows having missing values.

```python
#Solution to Q-6
c = 0

for x in (df.isnull()).sum(axis=1):
    if x!=0:
        c+=1
print(c)
```

Q-7: Discard the records with the missing values.

```python
#Solution to Q-7
ndf = df.dropna()
print(ndf.to_string())
```

 Q-8: Find the number of records after discarding null values.

```python
#Solution to Q-8
print(len(ndf.index))
```

Q-9: Replace the missing value with the mean/median/mode as applicable. 

```python
#Solution to Q-9
x = df["sepal length in cm"].mean()
df["sepal length in cm"].fillna(x, inplace = True)

x = df["sepal width in cm"].mode()
df["sepal width in cm"].fillna(x[0], inplace = True)

x = df["petal length in cm"].median()
df["petal length in cm"].fillna(x, inplace = True)

#and so on....
print(df.to_string())
```
