<b>Q-1: Create a dataframe and label the columns.

<b>Q-3:Normalize the dataset using z-score normalization

```python
#solution to q-3
zsdf = df.copy()
for feature_name in zsdf.columns[:4]:
    mean = zsdf[feature_name].mean()
    std = zsdf[feature_name].std()
    zsdf[feature_name] = (zsdf[feature_name] - mean) / std
print(zsdf.to_string())
```

<b>Q-4:Find the outliers using z-score normalization

```python
#solution to q-4
for index, row in zsdf.iterrows():
    for val in row[:4]:
        if(abs(val)>3):
            print(row)
            continue
```

<b> Q-5: Normalize the dataset using min-max normalization

```python
#solution to q-5
mmdf = df.copy()
for feature_name in mmdf.columns[:4]:
    max_value = mmdf[feature_name].max()
    min_value = mmdf[feature_name].min()
    mmdf[feature_name] = (mmdf[feature_name] - min_value) / (max_value - min_value)
mmdf
```

<b>Q-6:Write the code to discritize an attribute into 6 bins (third attribute)

```python
#solution to q-6
disdf = df.copy()
disdf["petal length"]=pd.cut(disdf["petal length"], bins=6)
print(disdf.to_string())
```
